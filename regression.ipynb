```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
model = LinearRegression().fit(x, y)
model.score(x,y)
```

```python
print(f"intercept: {model.intercept_}")


print(f"slope: {model.coef_}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the line equation: y = mx + b
def line_equation(x, m, b):
    return m * x + b

def plot_line(m, b, x_values, y_values):
  # Example values for the slope (m) and intercept (b)

  # Generate x values

  # Calculate y values using the line equation
  y_line = line_equation(x_values, m, b)

  # Plot the line
  plt.figure(figsize=(8, 6))
  plt.plot(x_values, y_line, label=f'y = {m}x + {b}', color='blue')

  plt.scatter(x_values, y_values, color='red', label='Points', zorder=5)


  # Add labels and title
  plt.title('Line Plot of y = mx + b', fontsize=14)
  plt.xlabel('x', fontsize=12)
  plt.ylabel('y', fontsize=12)
  plt.grid(True)

  # Add a legend
  plt.legend()

  # Show the plot
  plt.show()
```

```python
plt.plot(x, y, 'o')
```

```python
m = model.coef_[0]
b = model.intercept_
plot_line(m, b, x, y)
```
